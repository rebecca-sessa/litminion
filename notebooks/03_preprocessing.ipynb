```python
import litminion as lm
```

```python
lm.set_email("rebecca.sessa@studio.unibo.it")
```

```python
#download corpus

df = lm.download_pubmed(
    query="JAK inhibitor",
    max_results=10,
)
```

```python
#instantiate preprocessor

preprocessor = lm.ClassicalPreprocessor()
```

```python
#take single abstract

processed = preprocessor.transform(
    df.loc[0, "Abstract"]
)

print(processed)
```

```python
#process abstract

processed_documents = [
    preprocessor.transform(text)
    for text in df["Abstract"]
]
```

```python
#compare original vs processed

print(df.loc[0, "Abstract"][:500])

print("-" * 80)

print(processed_documents[0])
```
