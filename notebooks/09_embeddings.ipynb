```python
import litminion as lm
import numpy as np
```

```python
lm.set_email("rebecca.sessa@studio.unibo.it")
```

```python
corpus = lm.Corpus.from_pubmed(
    query="JAK inhibitor",
    max_results=25,
)

corpus
```

```python
preprocessor = lm.ClassicalPreprocessor()

corpus.preprocess(preprocessor)

corpus
```

```python
extractor = lm.SentenceTransformerExtractor()
```

```python
embeddings = extractor.fit_transform(
    corpus.get_documents()
)

type(embeddings)
```

```python
embeddings.shape
```

```python
extractor.get_embeddings().shape
```

```python
extractor.get_embedding_dimension()
```

```python
extractor.get_n_documents()
```

```python
embedding = extractor.get_embedding(0)

type(embedding), embedding.shape
```

```python
extractor.similarity(
    0,
    1,
)
```

```python
matrix = extractor.similarity_matrix()

matrix.shape
```

```python
matrix[:5,:5]
```

```python
results = extractor.most_similar(
    document=0,
    n=5,
)

results
```

```python
for idx, score in results:

    print("=" * 80)

    print(f"Document {idx}")
    print(f"Similarity: {score:.3f}\n")

    print(corpus.data.loc[idx, "Title"])
```

```python
results = extractor.search(
    query="JAK inhibitors for rheumatoid arthritis",
    n=5,
)

results
```

```python
for idx, score in results:

    print("=" * 80)

    print(f"Similarity: {score:.3f}\n")

    print(corpus.data.loc[idx, "Title"])
```
