```python
import litminion as lm
```

```python
#download + preprocess

lm.set_email("rebecca.sessa@studio.unibo.it")

df = lm.download_pubmed(
    query="JAK inhibitor",
    max_results=10,
)

preprocessor = lm.ClassicalPreprocessor()

processed_documents = [
    preprocessor.transform(text)
    for text in df["Abstract"]
]
```

```python
#instantiate analyzer

analyzer = lm.WordFrequencyAnalyzer()
```

```python
#fit

analyzer.fit(processed_documents)
```

```python
#vocab size

len(analyzer.transform())
```

```python
#top words

analyzer.top_n(20)
```

```python
#individual word frequency

counter = analyzer.transform()

counter["patient"]
counter["jak"]
counter["baricitinib"]
```
