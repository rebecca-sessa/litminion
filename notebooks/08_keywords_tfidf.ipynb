```python
import litminion as lm

lm.set_email("rebecca.sessa@studio.unibo.it")

corpus = lm.Corpus.from_pubmed(
    query="JAK inhibitor",
    max_results=50,
)

corpus.preprocess(
    lm.ClassicalPreprocessor()
)
```

```python
extractor = lm.TfidfKeywordExtractor()

extractor.fit(
    corpus.get_documents()
)
```

```python
extractor.get_keywords(20)
```

```python
extractor.to_dataframe(20)
```

```python
extractor.plot(20)
```
